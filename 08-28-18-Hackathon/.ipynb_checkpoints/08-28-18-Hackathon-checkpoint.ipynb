```python
train_df = pd.read_csv("train.csv")

# dropping unnamed columns
train_df = train_df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
print(train_df['price'])
# setting country to category and encoding
train_df['country'] = train_df['country'].astype('category')
train_df['country_code'] = train_df['country'].cat.codes

train_df = train_df[['price', 'country_code', 'points']].dropna()
#print(pd.notnull(train_df['price']))
#print(train_df['price'])
#train_df.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(random_state=0, n_jobs=-1)
regr.fit(train_df[['price', 'country_code']].values, train_df['points'].values)
print(regr.feature_importances_)
print(regr.predict([[34, 0]]))
```

```python
test_df = pd.read_csv("test.csv")

# dropping unnamed columns
test_df = test_df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
# setting country to category and encoding
test_df['country'] = test_df['country'].astype('category')
test_df['country_code'] = test_df['country'].cat.codes

test_df = test_df[['price', 'country_code', 'points']].dropna()
test_df.head()
len(test_df)
```

```python
regr.predict(test_df[['price', 'country_code']])
```

```python
from sklearn.model_selection import cross_val_score, KFold
kfold = KFold(n_splits=10, random_state=0)
results = cross_val_score(regr, test_df[['price', 'country_code']], test_df['points'], cv=kfold, scoring='r2')
results.mean()
```

```python
from sklearn.dummy import DummyRegressor
clf = DummyRegressor(strategy='mean')
clf.fit(train_df[['price', 'country_code']], train_df['points'])
clf.score(test_df[['price', 'country_code']], test_df['points'])
```

```python
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import r2_score

# Set the parameters by cross-validation
tuned_parameters = {'n_estimators': [500, 700, 1000], 'max_depth': [None, 1, 2, 3], 'min_samples_split': [2, 3]}

clf = GridSearchCV(RandomForestRegressor(verbose=1), tuned_parameters, scoring='r2', 
                   n_jobs=-1, verbose=1)
clf.fit(train_df[['price', 'country_code']], train_df['points'], )
print(clf.best_estimator_)
```
